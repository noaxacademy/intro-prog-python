#### **Cuaderno 3**: Calculando el área

Un problema muy frecuente en aplicaciones matemáticas y físicas es el cálculo de áreas, como por ejemplo, el área que queda debajo de una curva. 

A modo de ejemplo, en la siguiente figura se observa el área que queda debajo de la curva $f(x) = x^{1/2} + 1$, y delimitada por los extremos $x=2$ y $x=5$:

Área debajo de una curva

El problema de calcular el área bajo una curva se denomina integral. El caso presentado en la figura anterior se puede expresar como la siguiente integral:

$$ A = \int_2^5{(x^{1/2}+1)\,dx} $$

En funciones sencillas como esta, existen métodos analíticos para determinar la solución de la integral. Sin embargo, una aproximación al cálculo del área puede realizarse mediante la denominada __suma de Riemann__. Este procedimiento consiste en dividir el intervalo de integración (en este ejemplo, el intervalo $[2, 5]$ en varios subintervalos con el mismo «ancho», y aproximar el área bajo la curva sumando para ello diferentes rectángulos definidos por estos intervalos. 

Por ejemplo, si dividimos el intervalo en tres, obtenemos el siguiente escenario:

Suma de Riemann con tres rectángulos.

Se puede observar que la aproximación del área es bastante buena, sin embargo, es menor que el área real (de ahí los «huecos blancos» que quedan entre la parte superior de los rectángulos y la curva).

No obstante, es útil resaltar que esta aproximación será mejor cuantos más rectángulos hagamos caber en el intervalo de integración. Por ejemplo, si en lugar de tres subintervalos construimos seis, el resultado es el siguiente:

Suma de Riemann con seis rectángulos.

En este caso, los «huecos blancos» son menores, y la aproximación al área real es más precisa.
